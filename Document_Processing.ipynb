```python
pass_image=cv2.imread(r"/content/drive/MyDrive/sample-indian-passport-1.jpg")
cv2_imshow(pass_image)
```

Change the colour image to gray scale for more precise results

```python
pass_gray_image = cv.cvtColor(pass_image, cv.COLOR_BGR2GRAY)
thresh, thresh5_image=cv2.threshold(pass_gray_image, 200,255, cv2.THRESH_BINARY)
cv2_imshow(thresh5_image)
```

```python
doc2_image=cv2.imread(r"/content/drive/MyDrive/receipt.jpg")
cv2_imshow(doc2_image)
```

```python
doc2_gray_image = cv.cvtColor(doc2_image, cv.COLOR_BGR2GRAY)
thresh, thresh6_image=cv2.threshold(doc2_gray_image, 200,255, cv2.THRESH_BINARY)
cv2_imshow(thresh6_image)
```

**Truncated Thresholding**

Here there is no max value

If any pixel values is more than threshold value then its replaced by threshold value

If any pixel values is less than threshold value then pixel value is as it is

```python
thresh, thresh7_image=cv2.threshold(doc2_image, 235,255, cv2.THRESH_TRUNC)
cv2_imshow(thresh7_image)
```

**To Zero**

If any pixel values is more than threshold value then its maintained as it is

If any pixel values is less than threshold value then pixel value is replced by 0

```python
thresh, thresh8_image=cv2.threshold(doc2_image, 200,255, cv2.THRESH_TOZERO)
cv2_imshow(thresh8_image)
```

**To Zero Inverted**

If any pixel values is more than threshold value then pixel value is replced by 0

If any pixel values is less than threshold value then pixel value is maintained as it is

```python
thresh, thresh8_image=cv2.threshold(doc2_image, 200,255, cv2.THRESH_TOZERO_INV)
cv2_imshow(thresh8_image)
```

#Adaptive Thresholding

Types of Adaptive Thresholding : Mean Thresholding and Gaussian Thresholding

It makes thresholding adaptive ie., threshold value is automatically calculated

 Block size - size of neighbouring pixels

**Mean Thresolding** 

1.   Mean of block size(matrix) is taken by the adaptive
2.   This mean is taken as thresold value
3.   Apply a threshold required on this block size
4.   Take another block size and apply above three steps for it

```python
pass_thresh_image=cv2.adaptiveThreshold(pass_gray_image, 255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 21, 10)
#parameters:
#1. image, 2.max value for binary thresholding, 3.the kind of adaptive threshold used, 4. type of thresholding, 5. BLOCK SIZE mostly 21 is used for it,  6. applies weightage to the thresholding
cv2_imshow(pass_thresh_image)
```
