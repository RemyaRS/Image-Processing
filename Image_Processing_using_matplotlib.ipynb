In BGR format.... red colur is read as blue and blue colour is read as red colour...this problem needed to be solved

It can be solved by using convert colour function known as cvtColor

This fuction can be used to convert 

BGR --->RGB

BGR ---> GRAY

Also it can be used to convert format code

```python
rgb_image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
rgb_image
```

```python
plt.imshow(rgb_image)
```
